## Merged DataFrames

```python
#California and Wisconsin
merged_milk = pd.merge(milk_data_CA, milk_data_WI, on="year")
merged_milk.head()
```

```python
merged_milk = merged_milk.rename(columns={"Value_x": "CA Gross Income", "year":"Year", "Value_y": "WI Gross Income"})

merged_milk.head()
```

```python
#IDAHO AND NEW YORK
merged_milk2 = pd.merge(milk_data_ID, milk_data_NY, on="year")
merged_milk2.head()
```

```python
merged_milk2 = merged_milk2.rename(columns={"Value_x": "ID Gross Income", "year":"Year", "Value_y": "NY Gross Income"})

merged_milk2.head()
```

```python
#Merged California and Wisconsin
merged_milk_clean = merged_milk.drop(columns=['unit_desc_x', 'unit_desc_y', 'statisticcat_desc_x', 'statisticcat_desc_y'])
merged_milk_clean
```

```python
#Merged Idaho and New York
merged_milk_clean2 = merged_milk2.drop(columns=['unit_desc_x', 'unit_desc_y', 'statisticcat_desc_x', 'statisticcat_desc_y'])
merged_milk_clean2
```

```python
#Merging CA, WI, ID, and NY
Merged_df = pd.merge(merged_milk_clean, merged_milk_clean2, on="Year")
Merged_df
```

```python
#Cleaning TX DataFrame
milk_data_TX.head()
milk_data_TX = milk_data_TX.rename(columns={"Value": "TX Gross Income", "year": "Year"})
milk_data_TX
```

```python
#Merge TX into previous 4-state DataFrame and drop columns
Merged_Final = pd.merge(Merged_df, milk_data_TX, on="Year")

Merged_Final_df = Merged_Final.drop(columns=["unit_desc", "statisticcat_desc"])

Merged_Final_df.set_index('Year', drop=True, append=False, inplace=False, verify_integrity=False)
```

```python
#Clean up the columns
Merged_Final_df['CA Gross Income'] = Merged_Final_df['CA Gross Income'].str.replace(',', '')
Merged_Final_df['WI Gross Income'] = Merged_Final_df['WI Gross Income'].str.replace(',', '')
Merged_Final_df['ID Gross Income'] = Merged_Final_df['ID Gross Income'].str.replace(',', '')
Merged_Final_df['NY Gross Income'] = Merged_Final_df['NY Gross Income'].str.replace(',', '')
Merged_Final_df['TX Gross Income'] = Merged_Final_df['TX Gross Income'].str.replace(',', '')

#Change Datatype, reorder column, and set index
Merged_Final_df = Merged_Final_df.astype(int)
Merged_Final_df = Merged_Final_df.sort_values(by = 'Year', axis=0, ascending=True)
Merged_Final_df = Merged_Final_df.set_index('Year', drop=True, append=False, inplace=False, verify_integrity=False)

Merged_Final_df.dtypes
```

```python
Merged_Final_df
```

```python
ax = Merged_Final_df[['CA Gross Income','WI Gross Income', 'NY Gross Income', 
                              'ID Gross Income','TX Gross Income']].plot(kind='bar', zorder=3, 
                                                                         title ="Gross Income from Milk in \n Top 5 Milk Producing States", 
                                                                         figsize=(15, 5), legend= True, fontsize=10)                
ax.grid(zorder=0)
ax.legend(["California", "Wisconsin", 'New York', 'Idaho', 'Texas'])
plt.ticklabel_format(axis='y', style = 'sci', scilimits = (9,9))
for tick in ax.get_xticklabels():
    tick.set_rotation(0)
ax.set_xlabel("Fiscal Year", fontsize=12)
ax.set_ylabel("Gross Income (Billions)", fontsize=12)

plt.show()

```
